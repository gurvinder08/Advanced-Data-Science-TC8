```python
# number of words in text
data['No. of words'] = data['Text'].apply(lambda x: len(set(str(x).split())))
data
```

```python
# No. of characters including spaces
data['No. of char'] = data['Text'].apply(lambda x: len(str(x)))
data
```

```python
# Number of sentences
data['No. of sentences'] = data['Text'].apply(lambda x: len(str(x).split('.')))
data
```

#### Feature Engineering Techniques:


##### Removing special characters

```python
# Removing punctuations
import string
print(f'Sample punctuations: {list(string.punctuation)[0:10]}')
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree

data['clean_text']= data['Text'].apply(lambda x:remove_punctuation(x))
data[['Text', 'clean_text']]
```

##### Lower Casing

```python
# Lowering the text
data['lower_text'] = data['clean_text'].apply(lambda x: x.lower())
data[['Text', 'clean_text', 'lower_text']]
```

##### Contractions expansion

```python
!pip install contractions
```

```python
import contractions

data['contraction_expansion'] = data['lower_text'].apply(lambda x: contractions.fix(x))
data[['lower_text', 'contraction_expansion']]
```

##### Tokenization

```python
# Tokenization
import re
def tokenization(text):
    tokens = text.split()
    return tokens

data['tokenized'] = data['contraction_expansion'].apply(lambda x: tokenization(x))
data[['contraction_expansion', 'tokenized']]
```

##### Removing stop words

```python
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output

data['no_stopwords']= data['tokenized'].apply(lambda x:remove_stopwords(x))
data[['tokenized','no_stopwords']]
```

##### Stemming

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()

def stemming(text):
  stem_text = [porter_stemmer.stem(word) for word in text]
  return stem_text

data['stemmed'] = data['no_stopwords'].apply(lambda x: stemming(x))
data[['no_stopwords', 'stemmed']]
```

##### Lemmantization

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
nltk.download('omw-1.4')

def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text

data['lemmatized']=data['no_stopwords'].apply(lambda x:lemmatizer(x))
data[['no_stopwords','lemmatized']]
```

```python
corpus = [x[0] for x in data.lemmatized]
corpus = data['lemmatized'].apply(str)
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_cv = v.fit_transform(corpus)
v.vocabulary_
```
